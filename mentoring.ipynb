##### 추천시스템
- 도메인을 먼저 결정.

- 서비스로 만드는 것은 쉽지 않음.
- 컨텐츠벡터화하고 저장.

- 검색 후 유사성 높은 걸 매칭해야 함.

### flow
- 의료데이터로 학습 진행
- 사용자 feautre(나이, 병력) 최대한 많이 뽑아내기 - by gpt
- 학습 모델에 넣어서 추천 운동 및 내용 뽑기
- 문장 생성 - by gpt
    - lang chain
- 완성된 문장을 챗봇으로 출력

- 사용자 취향 / 건강상태 반영 운동 프로그램 추천.
- 샘플 데이터를 항목별로 다양하게 만들어서 경우의 수를 다양하게 가져가야 함.
- 사용자들의 취향에 따라 필요한데 없는 프로그램의 유무까지 판단 가능.
